Для початку намальюємо два простих графіка в логарифмічному масштабі
\begin{equation}
\cfrac{100^n}{n!} \qquad \cfrac{n!}{100^n}
\end{equation}

```python
fig, ax = subplots(1,2)
rng = range(0,1001)
ax[0].plot(rng, list(map(lambda x:100**x/math.factorial(x),rng)))
ax[0].set_yscale('log')
ax[0].set_xlabel('n')
rng = range(0,101)
ax[1].plot(rng, list(map(lambda x:math.factorial(x)/100**x,rng)))
ax[1].set_yscale('log')
ax[1].set_xlabel('n')
tight_layout()
show()
```

Намалюємо графік для функції 
\begin{equation}
F_n(n) = \sum\limits_{k=1}^n \cfrac{100^k}{k!}
\end{equation}
k від 1 до 2000, n від до 300

```python
fig, ay = subplots(3,1)
n_rng = range(1,301)
k_rng = range(1,2001)
def f_n(n):
    return reduce(lambda x,y:x+y, map(lambda k:100**k/math.factorial(k), range(1,n+1)))
ay[0].plot(n_rng, list(map(lambda x:f_n(x),n_rng)))
ay[0].set_yscale('log')
ay[1].plot(k_rng, list(map(lambda x:100**x/math.factorial(x),k_rng)))
k_rng = range(1,100) # переповнення 
ay[2].plot(k_rng, list(map(lambda x:math.factorial(x)/100**x,k_rng)))
ay[1].set_yscale('log')
ay[2].set_yscale('log')
show()
```

Намалюємо графік для функції 
\begin{equation}
F(n, w) = \sum\limits_{k=1}^n \cfrac{(-1)^k * k!}{w^k}
\end{equation}
w від 0.5 до 50 з кроком 0.5, f={10,50,100,130}

```python
def f_n_w(n,w):
    return reduce(lambda x,y:x+y, map(lambda k:(-1) ** k * math.factorial(k)/(w ** k), range(0,n+1)))
w_rng = np.arange(0.5,50.5,0.5)
plot(w_rng, list(map(lambda x:f_n_w(10,x),w_rng)),'b')
plot(w_rng, list(map(lambda x:f_n_w(50,x),w_rng)),'k')
plot(w_rng, list(map(lambda x:f_n_w(130,x),w_rng)),'r')
plot(w_rng, list(map(lambda x:f_n_w(100,x),w_rng)),'g')
yscale('log')
show() #Обрізати тут значення до 10
```

Намалюємо графік для функції 
\begin{equation}
F(n, w) = \int\limits_{k=0}^n \cfrac{w e^{-k}}{w+k} dx 
\end{equation}
w від 0.5 до 50 з кроком 0.5, f={10,50,100,130}

```python
def s_f_n_w(w,n):
    return quad(lambda x:w * math.exp(-x)/ (w+x), 0, n)[0]
w_rng = np.arange(0.5,50.5,0.5)
plot(w_rng, list(map(lambda x:s_f_n_w(x,10), w_rng)),'k')
plot(w_rng, list(map(lambda x:s_f_n_w(x,50), w_rng)),'b')
plot(w_rng, list(map(lambda x:s_f_n_w(x,130), w_rng)),'g')
plot(w_rng, list(map(lambda x:s_f_n_w(x,100), w_rng)),'r')
show()
```
